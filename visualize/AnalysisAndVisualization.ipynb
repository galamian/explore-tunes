# Some Visualizations


### Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import dataframe

```python
filepath = "../data/processed/tunes_1_to_202030.pkl"

df = pd.read_pickle(filepath)
```

### Tune ID distribution

```python
df["tune_id"] = pd.to_numeric(df["tune_id"])
binwidth = 500
df.hist('tune_id', bins=range(min(df["tune_id"]), max(df["tune_id"]) + binwidth, binwidth))
#df.plot(x='K', kind='bar')

#

plt.show()
```

### Key Distribution

```python
keys = df.groupby('K')['tune_id'].count().sort_values(ascending=False)
#keys = keys[keys>1000]

top = 15
print('Top ' + str(top) + ' keys in my dataset:')
print(keys[:top])
keys[:top].plot.bar()
plt.show()
```

### 'Rhythm' Distribution

```python
ry = df.groupby('R')['tune_id'].count().sort_values(ascending=False)

print('Top ' + str(top) + ' rhythms in my dataset:')
print(ry[:top])
ry[:top].plot.bar()
plt.show()
```
